```python
sh = pandas.read_csv('data.csv', header=None, skiprows=1, delimiter=';')
```

```python
N = 100
x = sh[2][:N]
y = sh[3][:N]
# mass = [ t/20203032 for t in sh[11][:N] ]
mass = sh[11][:N]
# z = np.asarray(mass)

# utmcoord = []
utmcoord = np.zeros((N, 2))
# z = []
# mass = []

bx, by = utm.from_latlon(x[0],y[0])[0:2]
bz = np.hypot(bx, by)
for i in range(len(x)):
    n, e =  utm.from_latlon(x[i],y[i])[0:2]
#     mass.append(np.hypot(n, e) - bz)
#     print(n,e)
#     utmcoord.append((n, e))
    utmcoord[i] = (n-bx, e-by)

z = np.asarray(mass)

# print(utmcoord)
# print(utmcoord,"\n",z)
```

```python
# np.linspace(min(x), max(x), num=N, endpoint=True)
X = np.linspace(min(x), max(x), endpoint=True)
Y = np.linspace(min(y), max(y), endpoint=True)
X, Y = np.meshgrid(X, Y)  # 2D grid for interpolation

# print()
```

```python
interp = LinearNDInterpolator(utmcoord, z)
Z = interp(X, Y)

plt.pcolormesh(X, Y, Z, shading='auto')
plt.plot(x, y, "ok", label="input point")
plt.legend()
plt.colorbar()
plt.axis("equal")
plt.show()
```

```python
# online example

rng = np.random.default_rng()
N = 500
subfac = 0.5
x = rng.random(N) - subfac
y = rng.random(N) - subfac
z = np.hypot(x, y)

X = np.linspace(min(x), max(x))
Y = np.linspace(min(y), max(y))

X, Y = np.meshgrid(X, Y)  # 2D grid for interpolation
# print(X.shape,"\n\n\n",Y.shape)
interp = LinearNDInterpolator(list(zip(x, y)), z)
Z = interp(X, Y)
# print(Z.shape)

plt.pcolormesh(X, Y, Z, shading='auto')
plt.plot(x, y, "ok", label="input point")
plt.legend()
plt.colorbar()
plt.axis("equal")
plt.show()
```
