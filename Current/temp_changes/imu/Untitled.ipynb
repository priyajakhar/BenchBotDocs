```python
files = ["gyrox.txt", "gyroy.txt", "gyroz.txt"]

for file in files:
    with open(file, "r") as my_file:
        data = my_file.read()
    data_into_list = data.replace('\n', ' ').split(" ")
    
    lst = []
    for x in data_into_list:
        if x == '':
            continue
        lst.append(float(x))

    plt.plot(lst)
    plt.show()
```
